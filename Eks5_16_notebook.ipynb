## Eksempel 16

Et tetraeder med hjørner i $A(1,-1,2), B(4,1,6), C(2,2,2)$ og $D(3,0,-2)$.<br>
Finn volumet av tetraederet.

```python
import numpy as np
A=np.array([1,-1,2])
B=np.array([4,1,6])
C=np.array([2,2,2])
D=np.array([3,0,-2])

AB=B-A
AC=C-A
AD=D-A

ABXAC=np.cross(AB,AC)
Volum=np.linalg.norm(np.dot(ABXAC,AD))
Volum=Volum/6
print(f"Volumet er {Volum}")
```
