## Hvordan finne ut om to vektorer er parallellelle
Noen av dere syntes det var vanskelig å forstå om to vektorer er parallelle enn "å bare se det". <br>
Det finnes flere måter, skalarproduktet eller lagranges identitet, men det enkleste (bortsett fra å se det) er vektorproduktet. <br>
Vi har to vektorer (som jeg vet er parallelle) $\vec{u}=[a,b,c]$ og $\vec{v}=[a\cdot k,b\cdot k,c\cdot k]$, der $k \in \mathbb{R}$, hvis vektor produktet er lik $\vec{0}$ er de to vektortene parallelle, under følger en utregning.

$
\vec{u}\times \vec{v} =\begin{bmatrix}
    \vec{e_x}       & \vec{e_y} &  \vec{e_z} \\
    a        & b &  c \\
    a\cdot k       & b\cdot k  &  c\cdot k
\end{bmatrix}=
\vec{e_x}
\begin{bmatrix} b & c\\
b\cdot k  &  c\cdot k
\end{bmatrix} - \vec{e_y}
\begin{bmatrix} a & c\\
a\cdot k  &  c\cdot k
\end{bmatrix}+\vec{e_z}\begin{bmatrix} a & b\\
a\cdot k  &  b\cdot k
\end{bmatrix}$ <br>
$\vec{u}\times \vec{v} =\vec{e_x} (bc\cdot k-bc\cdot k)-\vec{e_y}(ac\cdot k-ac\cdot k)+\vec{e_z}(ab\cdot k-ab\cdot k)$ <br>
$\vec{u}\times \vec{v} =[0,0,0]$

Vektorene $\vec{u}$ og $\vec{v}$ er dermed parallelle
