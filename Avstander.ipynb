## Eksempel 30

```python
import numpy as np

r=np.array([2,1,-2])
P=np.array([2,0,1])
Q=np.array([4,5,1])

def avstand_punkt_til_linje(P,Q,r):
    a=P-Q
    return np.linalg.norm(np.cross(a,r))/np.linalg.norm(r)

print(avstand_punkt_til_linje(P,Q,r))
```

### Felles metode for linje-linje og linje-punkt

```python
import numpy as np

r_l=np.array([-2,1,3])
r_m=np.array([2,-1,-3])
P=np.array([1,2,0])
Q=np.array([-3,1,4])


def avstand_to_linjer(P,Q):
    a=P-Q
    r=np.cross(r_l,r_m)
    return np.abs(np.dot(a,r))/np.linalg.norm(r)

def avstand_linje_punkt(P,Q,r):
    a=P-Q
    return np.linalg.norm(np.cross(a,r))/np.linalg.norm(r)

if np.linalg.norm(np.cross(r_l,r_m))!=0:
    print("ikke paralell",avstand_to_linjer(P,Q))

else:
    print("paralell",avstand_linje_punkt(P,Q,r_l))
```
